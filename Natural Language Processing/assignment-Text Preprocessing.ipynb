```python
document = '''"Sachin Ramesh Tendulkar, born 24 April 1973) is an Indian former international cricketer who captained the Indian national team. He is regarded as one of the greatest batsmen in the history of cricket.[4]
He is the highest run scorer of all time in international cricket, and the only player to have scored one hundred international centuries, the first batsman to score a double century in a One Day International (ODI), 
the holder of the record for the most runs in both Test and ODI cricket, and as of 2021 the only player to score more than 30,000 runs in international cricket.[5] 
He was the only Indian cricketer included in an all-time Test World XI compiled in 2013 to mark the 150th anniversary of Wisden Cricketers' Almanack.[6][7]
He is affectionately known as "Little Master" or "Master Blaster".'''
```

```python
document = document.lower()
```

```python
print(document.split())
```

# Tokenization 

```python
import nltk

# one time
# nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
```

```python
tokens = word_tokenize(document)
print(tokens)
```

```python
len(tokens) # words, numbers, special charachters 
```

# Stopwords

```python
# one time
# nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
print(stop_words)
```

```python
# remove stopwords from the documents

clean_tokens = [t for t in tokens if t not in stop_words]
print(clean_tokens)
```

```python
len(clean_tokens)
```

## Punctuation

```python
from string import punctuation
punctuation
```

```python
clean_tokens = [t for t in clean_tokens if t not in punctuation]
print(clean_tokens)
```

```python
len(clean_tokens)
```

## Stemming

```python
stemmer = nltk.PorterStemmer()
```

```python
stemmatized_tokens = [stemmer.stem(t) for t in clean_tokens]
print(stemmatized_tokens)
```

```python
clean_doc = ' '.join(stemmatized_tokens)
print(clean_doc)
```

## Lemmatization

```python
# one time
#nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemma = WordNetLemmatizer()
```

```python
# on clean tokens
lemmatized_tokens = [lemma.lemmatize(t) for t in clean_tokens]
print(lemmatized_tokens)
```

```python
clean_doc = ' '.join(lemmatized_tokens)
print(clean_doc)
```

## Parts Of Speech Tagging

```python
# one time
#nltk.download('averaged_perceptron_tagger')

nltk.pos_tag(lemmatized_tokens)
```
