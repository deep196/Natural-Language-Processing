```python
import transformers
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis",model="cardiffnlp/twitter-roberta-base-sentiment")
classifier("I've been waiting for a HuggingFace course my whole life.")
```

```python
classifier("What a bad movie, waste of time and money, need 2x return")
```

```python
classifier([
    "I've been waiting for a HuggingFace course my whole life.", 
    "I hate this so much!"
])
```

Named Entity recognition

```python
ner = pipeline("ner", grouped_entities=True)
```

```python
ner("Sukamal das is a trainer at Itvedant at Thane branch")
```

```python
classifier = pipeline("zero-shot-classification")
```

```python
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

```python
classifier(
    "Ronaldo is the best player in the world",
    candidate_labels=["cricket", "football", "racing", "batminton", "hockey"],
)
```

```python
question_answerer = pipeline("question-answering")
```

```python
question_answerer(
    question="Where do I work?",
    context="My name is Sukamal and I work at Itvedant in Thane"
)
```

```python
question_answerer(
    question="What is my name ?",
    context="My name is Sukamal and I work at Itvedant in Thane"
)
```

```python
question_answerer(
    question="Where is Itvedant situated?",
    context="My name is Sukamal and I work at Itvedant in Thane"
)
```
